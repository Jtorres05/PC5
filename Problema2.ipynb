```python
#Primer reporte: Vinos mejor puntuados por continente:
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv("./data/winemag-data-130k-v2.csv", sep = ",")

# 1. Renombrar columnas
df.rename(columns={'country': 'pais', 'points': 'puntuacion', 'price': 'precio', 'province': 'provincia'}, inplace=True)

# 2. Crear una nueva columna para etiquetar los países según su continente
continente = {
    'Italy': 'Europa',
    'Portugal': 'Europa',
    'US': 'América del Norte',
    'Spain': 'Europa',
    'France': 'Europa',
    'Germany': 'Europa',
    'Argentina': 'América del Sur',
    'Chile': 'América del Sur',
    'Australia': 'Oceanía',
    'Austria': 'Europa',
    'South Africa': 'África',
    'New Zealand': 'Oceanía',
    'Israel': 'Asia',
    'Hungary': 'Europa',
    'Greece': 'Europa',
    'Romania': 'Europa',
    'Mexico': 'América del Norte',
    'Canada': 'América del Norte',
    'Turkey': 'Asia',
    'Czech Republic': 'Europa',
    'Slovenia': 'Europa',
    'Luxembourg': 'Europa',
    'Croatia': 'Europa',
    'Georgia': 'Europa',
    'Uruguay': 'América del Sur',
    'England': 'Europa',
    'Lebanon': 'Asia',
    'Serbia': 'Europa',
    'Brazil': 'América del Sur',
    'Moldova': 'Europa',
    'Morocco': 'África',
    'Peru': 'América del Sur',
    'India': 'Asia',
    'Bulgaria': 'Europa',
    'Cyprus': 'Europa',
    'Armenia': 'Europa',
    'Switzerland': 'Europa',
    'Bosnia and Herzegovina': 'Europa',
    'Ukraine': 'Europa',
    'Slovakia': 'Europa',
    'Macedonia': 'Europa'
}

df['continente'] = df['pais'].map(continente)

# 3. Crear una nueva columna para calcular el precio por punto
df['precio_por_punto'] = df['precio'] / df['puntuacion']

# 4. Crear una nueva columna para clasificar los vinos según su precio
def clasificar_precio(precio):
    if precio < 20:
        return 'Económico'
    elif 20 <= precio < 50:
        return 'Medio'
    else:
        return 'Alto'

df['clasificacion_precio'] = df['precio'].apply(clasificar_precio)
```

```python
# Primer Reporte: Vinos mejor puntuados por continente
mejores_puntuados_por_continente = df.groupby('continente').apply(lambda x: x.nlargest(1, 'puntuacion'))
mejores_puntuados_por_continente.to_csv('mejores_puntuados_por_continente.csv', index=False)
```

```python
# Segundo Reporte: Cantidad de vinos clasificados por precio
cantidad_vinos_por_clasificacion_precio = df['clasificacion_precio'].value_counts()
cantidad_vinos_por_clasificacion_precio.plot(kind='bar', xlabel='Clasificación de Precio', ylabel='Cantidad de Vinos')
```

```python
# Tercer Reporte: Precio promedio por punto de vino por continente
precio_promedio_por_punto_por_continente = df.groupby('continente')['precio_por_punto'].mean()
precio_promedio_por_punto_por_continente.plot(kind='bar', xlabel='Continente', ylabel='Precio Promedio por Punto')
```

```python
# Reporte 4: Promedio de precio de vino por país, ordenado de mejor a peor
promedio_precio_por_pais = df.groupby('pais')['precio'].mean().reset_index()
promedio_precio_por_pais_sorted = promedio_precio_por_pais.sort_values(by='precio', ascending=False)
promedio_precio_por_pais_sorted.to_excel('promedio_precio_por_pais.xlsx', index=False)

promedio_precio_por_pais_sorted.head()
```
